## Google GEMINI Model Intergration

```python
import os
from langchain.chat_models import init_chat_model

model = init_chat_model("google_genai:gemini-3.1-flash-lite")
response = model.invoke("Why Do parrots talk?")
response.content
```

# STREAMING


```python
model = ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite")
```

```python
model.invoke("Write 200 words paragrpah on Generative AI")
```

```python
for chunk in model.stream("Write me a 200 words paragrpah on Generative AI"):
    print(chunk.text, end="", flush=True)
```

```python
for chunk in model.stream("Why do parrots have colorful feathers?"):
    print(chunk.text, end="")
```

## Batch

```python
responses = model.batch([
    "How does birds fly?",
    "Write about Indian Flag",
    "Write about Cockroach Janta Party."
])

for response in responses:
    print(response)
```
